```python
pop_seoul = pd.read_excel('01. population_in_Seoul.xls')
pop_seoul.head()
```

```python
# 필요한 데이터만 추출

pop_seoul=pd.read_excel('01. population_in_Seoul.xls', header=2, usecols='B,D,G,J,N')
pop_seoul.head()
```

```python
# 필드명 정리
pop_seoul.rename(columns={pop_seoul.columns[0]: '구별', pop_seoul.columns[1]: '인구수', pop_seoul.columns[2]: '내국인', pop_seoul.columns[3]: '외국인', pop_seoul.columns[4]: 
'고령자'}, inplace=True)
pop_seoul.head()
```
